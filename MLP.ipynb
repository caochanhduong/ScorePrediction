```python
import xlrd
import numpy as np
file_location = "dataOfficial.xlsx"
wb = xlrd.open_workbook(file_location)
sheet = wb.sheet_by_index(0)

print ("So dong:")
print (sheet.nrows)
print ("\nSo cot:")
print (sheet.ncols)
x_toan_train=np.zeros(sheet.nrows-1-100)#dữ liệu x, là điểm tổng kết
x_ly_train=np.zeros(sheet.nrows-1-100)
x_hoa_train=np.zeros(sheet.nrows-1-100)
x_sinh_train=np.zeros(sheet.nrows-1-100)
x_van_train=np.zeros(sheet.nrows-1-100)
x_su_train=np.zeros(sheet.nrows-1-100)
x_dia_train=np.zeros(sheet.nrows-1-100)
x_ngoaingu_train=np.zeros(sheet.nrows-1-100)
x_gdcd_train=np.zeros(sheet.nrows-1-100)

y_toan_train=np.zeros(sheet.nrows-1-100)#dữ liệu y, là điểm thi Tốt nghiệp
y_ly_train=np.zeros(sheet.nrows-1-100)
y_hoa_train=np.zeros(sheet.nrows-1-100)
y_sinh_train=np.zeros(sheet.nrows-1-100)
y_van_train=np.zeros(sheet.nrows-1-100)
y_su_train=np.zeros(sheet.nrows-1-100)
y_dia_train=np.zeros(sheet.nrows-1-100)
y_ngoaingu_train=np.zeros(sheet.nrows-1-100)
y_gdcd_train=np.zeros(sheet.nrows-1-100)


x_toan_test=np.zeros(100)#dữ liệu x, là điểm tổng kết
x_ly_test=np.zeros(100)
x_hoa_test=np.zeros(100)
x_sinh_test=np.zeros(100)
x_van_test=np.zeros(100)
x_su_test=np.zeros(100)
x_dia_test=np.zeros(100)
x_ngoaingu_test=np.zeros(100)
x_gdcd_test=np.zeros(100)

y_toan_test=np.zeros(100)#dữ liệu y, là điểm thi Tốt nghiệp
y_ly_test=np.zeros(100)
y_hoa_test=np.zeros(100)
y_sinh_test=np.zeros(100)
y_van_test=np.zeros(100)
y_su_test=np.zeros(100)
y_dia_test=np.zeros(100)
y_ngoaingu_test=np.zeros(100)
y_gdcd_test=np.zeros(100)

#đọc và chia dữ liệu train test cho 9 môn thi 
print ("\nCac gia tri trong cot :")
for x in range(sheet.ncols):
    print ("\n")
    for rows in range(sheet.nrows):
        if x==0 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_toan_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_toan_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==1 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_ly_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_ly_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==2 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_hoa_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_hoa_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==3 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_sinh_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_sinh_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==4 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_van_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_van_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==5 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_su_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_su_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==6 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_dia_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_dia_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==7 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_ngoaingu_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_ngoaingu_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==8 and rows>=1:
            if rows<= sheet.nrows-1-100:
                x_gdcd_train[rows-1]=sheet.cell_value(rows, x)
            else:
                x_gdcd_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
            
        if x==12 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_toan_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_toan_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==13 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_ly_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_ly_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==14 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_hoa_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_hoa_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==15 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_sinh_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_sinh_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==16 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_van_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_van_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==17 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_su_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_su_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==18 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_dia_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_dia_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==19 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_ngoaingu_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_ngoaingu_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)
        if x==20 and rows>=1:
            if rows<= sheet.nrows-1-100:
                y_gdcd_train[rows-1]=sheet.cell_value(rows, x)
            else:
                y_gdcd_test[rows-1-sheet.nrows+1+100]=sheet.cell_value(rows, x)

            
#in thử môn cột toán
print("toán train\n")
print(len(x_gdcd_train))

print("\n")
print("toán test \n")
print(len(y_gdcd_train))

    
```

```python
#đưa vào randomforestregressor
from sklearn.neural_network import MLPRegressor
regr_toan = MLPRegressor()
regr_ly = MLPRegressor()
regr_hoa = MLPRegressor()
regr_sinh = MLPRegressor()
regr_van = MLPRegressor()
regr_su = MLPRegressor()
regr_dia = MLPRegressor()
regr_ngoaingu = MLPRegressor()
regr_gdcd = MLPRegressor()

def round_quarter(x):
    return np.round_(x*4)/4
def round_fifth(x):
    return np.round_(x*4)/5

regr_toan.fit(x_toan_train.reshape(-1,1), y_toan_train.reshape(-1,1))
regr_ly.fit(x_ly_train.reshape(-1,1), y_ly_train.reshape(-1,1))
regr_hoa.fit(x_hoa_train.reshape(-1,1), y_hoa_train.reshape(-1,1))
regr_sinh.fit(x_sinh_train.reshape(-1,1), y_sinh_train.reshape(-1,1))
regr_van.fit(x_van_train.reshape(-1,1), y_van_train.reshape(-1,1))
regr_su.fit(x_su_train.reshape(-1,1), y_su_train.reshape(-1,1))
regr_dia.fit(x_dia_train.reshape(-1,1), y_dia_train.reshape(-1,1))
regr_ngoaingu.fit(x_ngoaingu_train.reshape(-1,1), y_ngoaingu_train.reshape(-1,1))
regr_gdcd.fit(x_gdcd_train.reshape(-1,1), y_gdcd_train.reshape(-1,1))



score_toan=regr_toan.score(x_toan_test.reshape(-1,1), y_toan_test.reshape(-1,1))
score_ly=regr_ly.score(x_ly_test.reshape(-1,1), y_ly_test.reshape(-1,1))
score_hoa=regr_hoa.score(x_hoa_test.reshape(-1,1), y_hoa_test.reshape(-1,1))
score_sinh=regr_sinh.score(x_sinh_test.reshape(-1,1), y_sinh_test.reshape(-1,1))
score_van=regr_van.score(x_van_test.reshape(-1,1), y_van_test.reshape(-1,1))
score_su=regr_su.score(x_su_test.reshape(-1,1), y_su_test.reshape(-1,1))
score_dia=regr_dia.score(x_dia_test.reshape(-1,1), y_dia_test.reshape(-1,1))
score_ngoaingu=regr_ngoaingu.score(x_ngoaingu_test.reshape(-1,1), y_ngoaingu_test.reshape(-1,1))
score_gdcd=regr_gdcd.score(x_gdcd_test.reshape(-1,1), y_gdcd_test.reshape(-1,1))
print("Độ chính xác môn toán: \n",format(score_toan, '.2f'))
print("Độ chính xác môn lý: \n",format(score_ly, '.2f'))
print("Độ chính xác môn hoá: \n",format(score_hoa, '.2f'))
print("Độ chính xác môn sinh: \n",format(score_sinh, '.2f'))
print("Độ chính xác môn văn: \n",format(score_van, '.2f'))
print("Độ chính xác môn sử: \n",format(score_su, '.2f'))
print("Độ chính xác môn địa: \n",format(score_dia, '.2f'))
print("Độ chính xác môn ngoại ngữ: \n",format(score_ngoaingu, '.2f'))
print("Độ chính xác môn gdcd: \n",format(score_gdcd, '.2f'))
```
